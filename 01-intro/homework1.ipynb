Q7. Sum of weights

```python
X = df[df["origin"]=="Asia"][["vehicle_weight", "model_year"]].iloc[0:7, :].to_numpy()
```

```python
X
```

```python
X.T
```

```python
XTX = np.matmul(X.T, X)
```

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
```

```python
w = np.matmul(np.matmul(np.linalg.inv(XTX), X.T), y)
```

```python
w.sum()
```
